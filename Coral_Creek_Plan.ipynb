```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import yfinance as yf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LayerNormalization, Dropout, MultiHeadAttention, GlobalAveragePooling1D
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import math
import logging

# 使用已安装的中文字体
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Microsoft YaHei', 'STHeiti']  # 替换为系统中的字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 配置日志记录
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# 定义 Transformer 块
class TransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim),
        ])
        self.layernorm1 = LayerNormalization(epsilon=1e-6)
        self.layernorm2 = LayerNormalization(epsilon=1e-6)
        self.dropout1 = Dropout(rate)
        self.dropout2 = Dropout(rate)

    def call(self, inputs, training=False):
        attn_output = self.att(inputs, inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

# 定义位置嵌入
class TokenAndPositionEmbedding(tf.keras.layers.Layer):
    def __init__(self, maxlen, embed_dim):
        super(TokenAndPositionEmbedding, self).__init__()
        self.token_emb = Dense(embed_dim)
        self.pos_emb = tf.keras.layers.Embedding(input_dim=maxlen, output_dim=embed_dim)

    def call(self, x):
        maxlen = tf.shape(x)[-2]
        positions = tf.range(start=0, limit=maxlen, delta=1)
        positions = self.pos_emb(positions)
        x = self.token_emb(x)
        return x + positions

# 定义 Transformer 模型类
class TransformerModel:
    def __init__(self, ticker, start_date, end_date, target_column='Close', window_size=60, training_fraction=0.7, validation_fraction=0.15, epochs=100, batch_size=32, use_technical_indicators=False):
        """
        初始化 TransformerModel，包括股票代码、日期范围、训练参数和是否使用技术指标。
        """
        self.ticker = ticker
        self.start_date = start_date
        self.end_date = end_date
        self.target_column = target_column
        self.window_size = window_size
        self.training_fraction = training_fraction
        self.validation_fraction = validation_fraction
        self.epochs = epochs
        self.batch_size = batch_size
        self.use_technical_indicators = use_technical_indicators
        self.scaler = MinMaxScaler()
        self.close_scaler = MinMaxScaler()
        self.model = None
        self.predictions = None
        self.rmse = None
        self.mae = None
        self.mape = None

    def download_data(self):
        """
        使用 yfinance 下载股票数据。
        """
        logger.info(f"下载股票数据: {self.ticker} 从 {self.start_date} 到 {self.end_date}")
        data = yf.download(self.ticker, start=self.start_date, end=self.end_date)
        if data.empty:
            logger.error("下载的数据为空，请检查股票代码和日期范围。")
            raise ValueError("下载的数据为空，请检查股票代码和日期范围。")
        logger.info(f"数据下载完成，数据形状: {data.shape}")
        data.reset_index(inplace=True)
        return data

    def add_technical_indicators(self, data):
        """
        计算并添加常见的技术指标。
        """
        logger.info("计算技术指标...")
        # 添加技术指标
        data['SMA_30'] = data['Close'].rolling(window=30).mean()
        data['SMA_50'] = data['Close'].rolling(window=50).mean()
        data['SMA_200'] = data['Close'].rolling(window=200).mean()
        data['RSI'] = 100.0 - (100.0 / (1.0 + (data['Close'].diff().clip(lower=0).rolling(window=14).mean() / (-data['Close'].diff().clip(upper=0).rolling(window=14).mean()))))
        data['MACD'] = data['Close'].ewm(span=12, adjust=False).mean() - data['Close'].ewm(span=26, adjust=False).mean()
        data['MACDSignal'] = data['MACD'].ewm(span=9, adjust=False).mean()
        data['MACDHist'] = data['MACD'] - data['MACDSignal']
        logger.info("技术指标计算完成。")
        return data

    def preprocess_data(self, data):
        """
        执行数据预处理，包括计算技术指标（可选）、处理缺失值和缩放数据。
        """
        logger.info("开始数据预处理...")
        if self.use_technical_indicators:
            data = self.add_technical_indicators(data)
        else:
            logger.info("跳过技术指标计算，仅使用收盘价进行预测。")

        # 移除日期列，并确保所有数据都是数值类型
        data = data.drop(columns=['Date'])

        # 处理缺失值：前向填充和删除仍有缺失值的行
        data.fillna(method='ffill', inplace=True)
        data.dropna(inplace=True)
        logger.info(f"数据形状在处理缺失值后: {data.shape}")

        # 缩放数据
        feature_columns = [col for col in data.columns if col != 'Date']
        data[feature_columns] = self.scaler.fit_transform(data[feature_columns])

        # 可选：保存预处理后的数据以供检查
        data.to_csv('preprocessed_data.csv', index=False)
        logger.info("预处理后的数据已保存为 'preprocessed_data.csv'。")

        return data

    def create_sequences(self, data_scaled):
        """
        使用滑动窗口方法创建序列数据。
        """
        logger.info("创建序列数据...")
        scaled_array = data_scaled.values
        X, y = [], []
        for i in range(self.window_size, len(scaled_array)):
            X.append(scaled_array[i-self.window_size:i, :])
            y.append(scaled_array[i, 0])  # 目标是 'Close' 列
        X, y = np.array(X), np.array(y)
        logger.info(f"序列数据形状: X={X.shape}, y={y.shape}")
        return X, y

    def prepare_data_for_model(self, data):
        """
        准备模型训练所需的数据，包括缩放和序列创建。
        """
        logger.info("准备模型训练数据...")

        # 创建序列
        X, y = self.create_sequences(data)

        # 分割训练集、验证集和测试集
        training_data_len = int(len(X) * self.training_fraction)
        validation_data_len = int(len(X) * self.validation_fraction)

        X_train, y_train = X[:training_data_len], y[:training_data_len]
        X_val, y_val = X[training_data_len:training_data_len+validation_data_len], y[training_data_len:training_data_len+validation_data_len]
        X_test, y_test = X[training_data_len+validation_data_len:], y[training_data_len+validation_data_len:]

        logger.info(f"训练集样本数: {X_train.shape[0]}, 验证集样本数: {X_val.shape[0]}, 测试集样本数: {X_test.shape[0]}")

        return X_train, y_train, X_val, y_val, X_test, y_test

    def build_transformer_model(self, embed_dim=64, num_heads=4, ff_dim=128, dropout_rate=0.1):
        """
        构建 Transformer 模型架构。
        """
        logger.info("构建 Transformer 模型...")
        input_shape = self.X_train.shape[1], self.X_train.shape[2]
        inputs = Input(shape=input_shape)
        embedding_layer = TokenAndPositionEmbedding(self.window_size, embed_dim)(inputs)
        transformer_block = TransformerBlock(embed_dim, num_heads, ff_dim, dropout_rate)(embedding_layer)
        pooling = GlobalAveragePooling1D()(transformer_block)
        dropout = Dropout(0.1)(pooling)
        outputs = Dense(1)(dropout)

        self.model = Model(inputs=inputs, outputs=outputs)
        self.model.compile(optimizer='adam', loss='mean_squared_error')
        logger.info("Transformer 模型已构建和编译。")
        self.model.summary()

    def train_transformer_model(self):
        """
        训练 Transformer 模型，使用早停和学习率调度器。
        """
        logger.info("开始训练 Transformer 模型...")
        early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
        reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)
        history = self.model.fit(
            self.X_train, self.y_train,
            validation_data=(self.X_val, self.y_val),
            epochs=self.epochs,
            batch_size=self.batch_size,
            callbacks=[early_stop, reduce_lr],
            verbose=1
        )
        logger.info("模型训练完成。")

        # 绘制训练和验证损失曲线
        plt.figure(figsize=(10,6))
        plt.plot(history.history['loss'], label='训练损失')
        plt.plot(history.history['val_loss'], label='验证损失')
        plt.legend()
        plt.xlabel('Epoch')
        plt.ylabel('损失 (MSE)')
        plt.title('Transformer 模型损失曲线')
        plt.grid()
        plt.show()

    def predict_and_evaluate(self):
        """
        使用训练好的模型进行预测，并评估模型性能。
        """
        logger.info("使用测试集进行预测...")
        predictions = self.model.predict(self.X_test)

        # 逆缩放预测值和实际值
        logger.info("逆缩放预测值和实际值...")
        predictions_inverse = self.close_scaler.inverse_transform(predictions).flatten()
        y_test_inverse = self.close_scaler.inverse_transform(self.y_test.reshape(-1, 1)).flatten()

        self.predictions = predictions_inverse
        self.y_test_inverse = y_test_inverse

        # 计算评估指标
        self.rmse = np.sqrt(mean_squared_error(y_test_inverse, self.predictions))
        self.mae = mean_absolute_error(y_test_inverse, self.predictions)
        self.mape = mean_absolute_percentage_error(y_test_inverse, self.predictions)
        logger.info(f'Transformer 模型评估指标:')
        logger.info(f'MAE: {self.mae}')
        logger.info(f'RMSE: {self.rmse}')
        logger.info(f'MAPE: {self.mape}')

    def plot_predictions(self):
        """
        绘制实际值与预测值的对比图。
        """
        logger.info("绘制实际值与预测值的对比图...")
        plt.figure(figsize=(14,7))
        plt.plot(self.y_test_inverse, label='实际收盘价', color='blue')
        plt.plot(self.predictions, label='预测收盘价', color='orange')
        plt.legend()
        plt.title('Transformer 预测值与实际值对比')
        plt.xlabel('时间')
        plt.ylabel('价格 (USD)')
        plt.grid()
        plt.show()

    def run(self):
        """
        运行整个预测流程。
        """
        try:
            # 下载数据
            data = self.download_data()

            # 预处理数据
            data = self.preprocess_data(data)

            # 准备模型训练数据
            X_train, y_train, X_val, y_val, X_test, y_test = self.prepare_data_for_model(data)
            self.X_train, self.y_train = X_train, y_train
            self.X_val, self.y_val = X_val, y_val
            self.X_test, self.y_test = X_test, y_test

            # 构建模型
            self.build_transformer_model()

            # 训练模型
            self.train_transformer_model()

            # 预测与评估
            self.predict_and_evaluate()

            # 绘制预测结果
            self.plot_predictions()

            return self.rmse, self.mae, self.mape
        except Exception as e:
            logger.error(f"运行过程中出现错误: {e}")
            raise

# 定义函数来运行并比较两个模型
def compare_models():
    # 设置参数
    ticker = 'TSLA'  # 股票代码，可以更改为其他股票
    start_date = '2010-01-01'  # 开始日期
    end_date = '2023-12-31'  # 结束日期

    # 模型1：仅使用收盘价
    logger.info("=========== 运行模型1：仅使用收盘价 ===========")
    transformer_price_only = TransformerModel(
        ticker=ticker,
        start_date=start_date,
        end_date=end_date,
        target_column='Close',
        window_size=60,
        training_fraction=0.7,
        validation_fraction=0.15,
        epochs=100,
        batch_size=32,
        use_technical_indicators=False  # 不使用技术指标
    )
    try:
        rmse1, mae1, mape1 = transformer_price_only.run()
        logger.info(f"模型1 - 评估指标:\nRMSE: {rmse1}\nMAE: {mae1}\nMAPE: {mape1}\n")
    except Exception as e:
        logger.error(f"模型1 运行过程中出现错误: {e}")

    # 模型2：使用多种技术指标
    logger.info("=========== 运行模型2：使用多种技术指标 ===========")
    transformer_with_features = TransformerModel(
        ticker=ticker,
        start_date=start_date,
        end_date=end_date,
        target_column='Close',
        window_size=60,
        training_fraction=0.7,
        validation_fraction=0.15,
        epochs=100,
        batch_size=32,
        use_technical_indicators=True  # 使用技术指标
    )
    try:
        rmse2, mae2, mape2 = transformer_with_features.run()
        logger.info(f"模型2 - 评估指标:\nRMSE: {rmse2}\nMAE: {mae2}\nMAPE: {mape2}\n")
    except Exception as e:
        logger.error(f"模型2 运行过程中出现错误: {e}")

    # 对比结果
    logger.info("=========== 模型对比 ===========")
    comparison_df = pd.DataFrame({
        '模型': ['仅使用收盘价', '使用多种技术指标'],
        'RMSE': [rmse1, rmse2],
        'MAE': [mae1, mae2],
        'MAPE': [mape1, mape2]
    })
    print(comparison_df)

    # 可视化对比
    comparison_df_melted = comparison_df.melt(id_vars='模型', var_name='指标', value_name='值')
    plt.figure(figsize=(10,6))
    sns.barplot(x='指标', y='值', hue='模型', data=comparison_df_melted)
    plt.title('模型评估指标对比')
    plt.ylabel('值')
    plt.xlabel('评估指标')
    plt.legend(title='模型')
    plt.show()

if __name__ == "__main__":
    compare_models()
```
